```python
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#Sửa tên các cột 
df.columns = df.columns.str.lower() #đưa các tên cột về chữ cái thường
```

```python
df.columns = df.columns.str.replace('_' , '' )
df.columns = df.columns.str.replace('.' , '' )
df.columns = df.columns.str.replace('%' , '_' )
```

```python
df.rename(columns = {'inforreturn' : 'infor_return'} , inplace = True )
df.rename(columns = {'inforret' : 'infor_ret'} , inplace = True)
df.rename(columns = {'name_book' : 'book_name'} , inplace = True)
```

```python
df = df.drop(columns = ["author_id" , "author_name"] ,axis = 1 )  #Xóa các cột có giá trị null
```

```python
df.fillna(0 , inplace = True)
```

```python
# Chuyển đổi kiểu dữ liệu
df.info()
```

```python
df["quantity_sold"] = df["quantity_sold"].astype(int)
```

```python
df.iloc[400:500 , :]
```

```python
#Loại bỏ các thẻ HTML có trong cột description vaf quantity_sold 
def clean_html(text) :
    if text == 0 :  
        return "0"  #Nếu gặp giá trị 0 (là giá trị null trước đó) thì đổi trả về 0
    else :
        return BeautifulSoup(text , 'html.parser').get_text(separator = ' ').strip()
    
df['description'] = df['description'].apply(clean_html)
df['infor_return'] = df['infor_return'].apply(clean_html)
```
