```python
params = {
    'platform' : 'web' ,
    'spid' : 275463058 ,
    'version' : 3
}
```

```python
# Hàm lấy dữ liệu prouct
def get_product_data(json) : 
    d = {}
    d["iD"] = json.get('id')
    d["Sku"] = json.get('sku',None)
    d["Name%book_"] = json.get('name' , None)
    d["Origin%price."] = json.get('original_price' , None)
    d['Discount%rate'] = json.get('discount_rate',None) 
    d['discount']  = json.get('discount' , None)
    d["Price"] = json.get('price' , None)
    d['Description'] = json.get('description' , None)
    quantity_sold = json.get('quantity_sold')
    d['Quantity%sold___'] = quantity_sold.get('value') if isinstance(quantity_sold , dict) else None
    categories = json.get('categories')
    d['Categories'] = categories.get('name') if isinstance(categories , dict) else None
    authors = json.get('authors')
    d["Author%id"] = authors.get('id') if isinstance(authors , dict) else None
    d["Author%name___"] = authors.get('name') if isinstance(authors , dict) else None
    d["infor_return"] = json.get('return_and_exchange_policy' , None)
    d['gift%item%title'] = json.get('gift_item_title' , None)
    return d 
```

```python
product_id = pd.read_csv(r'D:\VS-workspace\Week1\dataset\product_id.csv')
prd_id = product_id.id.to_list() # Chuyển bảng id thành list
```

```python
product_data = []
```

```python
# Lấy thông tin product từ mỗi product_id trong list , dùng tqdm để quản lý tiến trình
for i in tqdm(prd_id , total = len(prd_id)) :
    response = requests.get("https://tiki.vn/api/v2/products/{}".format(i) , headers = headers , params = params , cookies = cookies ) # lấy api từ web tiki
    if response.status_code == 200 :
        try : 
            print("Crawl product_data {} success !!".format(i))
            product_data.append(get_product_data(response.json()))  #Lấy toàn bộ thông tin dict và thêm vào list product_data
        except ValueError : 
            print("Không thể parse JSON , respone không hợp lệ") 
    else :
        print("Request lỗi")
            
        
```

```python
df = pd.DataFrame(product_data)
```
